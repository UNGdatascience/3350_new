# Statistical Estimation

## Summary of the Caffeine Variable

Let’s consider a brief example where researchers wish to esimate the average **Coping Humor Usage** by students on the UNG Dahlonega campus drink per day. Let’s load, subset and summarize the correct vector of values:

```r
pers <- read.csv('https://faculty.ung.edu/rsinn/data/personality.csv')
cope <-pers[,'CHS']
cat ("Descriptive Statistics for Caffeine: \n  Mean =", round(mean(cope),2), "\n  Standard Deviation =", round(sd(cope),2), "\n  Sample Size =", length(cope),"\n\nThe 5-number summary is shown below.")
summary(cope)
```

### Shape

To better understand the shape of population from which the data were drawn, let’s plot a histogram and density plot.

```r
layout(matrix(c(1,2), ncol=2), height = lcm(12))
plot(density(cope), main = 'Density: Coping Humor')
plt <- {qqnorm(cope, main = 'QQ Plot: Coping Humor') ; qqline(cope)} 

hist(cope, main = 'Histogram: Coping Humor')
```

We find the distribution approximately normal with some quirkiness. The data are appropriate for 
-procedures, yet more pronounced quirks might force us to reject the data as inappropriate.

### Estimate
Question: Find a 95% confidence interval estimate for the Coping Humor Usage for UNG Dahlonega students using the *personality.csv* data set.

To answer, we can use the t.test function which returns the confidence interval for any test we run. Here, we choose so that the function will run, not for any statistical reason.

```r
t.test(cope, mu = 25)
```

**Answer:** The 95% confidence interval for the mean of the Coping Humor variable is between 
$24.3$ and $25.9$.

**Meaning:** Given repeated samples of the same size from the same population, the calcuated means of the samples will fall within the confidence interval 95\% of the time.

### Changing Confidence Level

If we wish to change the default setting of $95\%$, we can add the **conf.level** option to do so.

```r
t.test(cope, mu = 25, conf.level = 0.99)
```

## $z$-Proportion Confidence Intervals

If we are estimating the support for our condidate in an upcoming election and have polling data show that, in the Atlanta metro area, 512 likely voters will vote for her while 487 will vote against.

## Data

We must first create an observed matrix for the data collected which proceeds as follows:

```r
obs <- matrix(c(512,486),ncol=2)
obs
```

## Assumptions

The assumptions for confidence intervals are the same as the statistical tests using that variable. In this case, we need a reasonably large sample and proportion data which, by inspection, we clearly have.

## Estimate

We will be using the prop.test function because it performs $z$-proportion testing, and we wish to produce a $z$-proportion confidence interval.

```r
prop.test(obs)
```

## Estimate 

The 95% confidence interval for the percentage of voters likely to vote for our candidate is between $48.2\%$ and $54.4\%$.

As before, we can change the conf.level to $99\%$ as follows:

```r
prop.test(obs, conf.level = 0.99)
```
