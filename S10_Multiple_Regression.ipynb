# S10: Multiple Regression

The process mirrors linear bivariate regression which was covered in section 5 of these notes. The key difference is that we use multiple predictors to estimate the dependent variable.

## Example
We want to predict a penguin's Body Mass (g) using two independent variables:
- Flipper Length (mm)
- Bill Depth (mm)

### The Statistical Model

In multiple linear regression, we use the equation:$$y = \beta_0 + \beta_1x_1 + \beta_2x_2 + \epsilon$$

Where:
- $y$: Body Mass (Dependent variable)
- $x_1$: Flipper Length
- $x_2$: Bill Depth
- $\beta_0$: The Intercept
- $\beta_{1,2}$: The coefficients (slopes) for each variable

Low Multicollinearity
In many datasets (like real estate), the predictors move together. In this insurance data:
- Age vs. BMI: Being older doesn't strictly mean you have a higher BMI.
- Age vs. Children: While there is a slight lifecycle trend, people of the same age have widely varying numbers of children.
- BMI vs. Children: These are generally independent variables.

4. Interpretation of Results
When you look at the Coefficients (coef):
- Age: Typically shows a positive coefficient (around 239), meaning each year of age adds roughly $239 to the annual premium.
- BMI: Also positive; as BMI increases, the statistical risk and cost increase.
- Children: Each additional child generally increases the insurance charge.

The VIF scores for these three variables usually hover between 1.0 and 1.1 (anything under 5.0 is excellent). This means the information provided by "Age" is unique and not being "stolen" or overlapped by the information in "BMI" or "Children."
